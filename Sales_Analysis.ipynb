# Data Analysis Report

## Introduction

In this data analysis report, we explore and analyze the sales data from an online store. The dataset contains information about customer demographics, product categories, and sales figures. The goal of this analysis is to identify trends and patterns that can provide valuable insights for business optimization.

## Exploratory Data Analysis

The exploratory data analysis (EDA) phase involves understanding the dataset's structure, data types, and initial observations. We import the required libraries, load the data, and perform data cleaning and preprocessing tasks.


```python
# Importing Required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  # For visualizing data
%matplotlib inline
import seaborn as sns
```

```python
# Load the Data
data_file_path = 'Online_Sales_Data.csv'
df = pd.read_csv(data_file_path)
```

```python
# Getting the number of rows and columns in the DataFrame
num_rows, num_columns = df.shape
print("The DataFrame has {} rows and {} columns.".format(num_rows, num_columns))
```

```python
# Displaying the first few rows of the DataFrame

df.head()
```

```python
# Displaying the last few rows of the DataFrame

df.tail()
```

```python
# Replace the invalid characters "�" with an blank
df['State'] = df['State'].str.replace("�", " ")
df.head()
```

```python
# Getting descriptive statistics of the DataFrame

df.describe()
```

```python
# Displaying the concise summary of the DataFrame

df.info()
```

```python
# Dropping unrelated/blank columns from the DataFrame

columns_to_drop = ['Status', 'unnamed1']
df.drop(columns_to_drop, axis=1, inplace=True)
```

```python
# Checking for null (missing) values in the DataFrame

pd.isnull(df)
```

```python
# Checking for null (missing) values in the DataFrame and calculating the count of null values per column

df.isna().sum()
```

```python
# Dropping rows with null (missing) values from the DataFrame

df.dropna(inplace=True)
```

```python
# Changing the data type of the 'Amount' column to integer

df['Amount'] = df['Amount'].astype('int')
```

```python
# Getting the column names of the DataFrame

df.columns
```

```python
# Custom function to map age to bucket label
def get_age_bucket(age):
    bins = [0, 9, 19, 29, 39, 49, 59, 69, 79, float('inf')]
    labels = ["0-9", "10-19", "20-29", "30-39", "40-49", "50-59", "60-69", "70-79", "80+"]
    for i in range(len(bins) - 1):
        if age >= bins[i] and age < bins[i+1]:
            return labels[i]
    return "Unknown"

# Apply the function to create the Age_Buckets column
df['Age_Buckets'] = df['Age'].apply(get_age_bucket)

# Getting the count of unique values in the 'Age_Buckets' column
df['Age_Buckets'].value_counts()
```

## **Gender**

```python
# Creating a count plot to visualize the distribution of 'Gender'
ax = sns.countplot(x='Gender', data=df)

# Adding labels to the bars with their respective counts
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Adding plot labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Gender in the Dataset')

# Displaying the plot
plt.show()
```

```python
# Grouping the DataFrame by 'Gender', calculating the sum of 'Amount', and sorting the results
gender_amount_sum = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Displaying the results
print(gender_amount_sum)
```

```python
# Grouping the DataFrame by 'Gender', calculating the sum of 'Amount', and sorting the results
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Creating a bar plot to visualize the total sales amount for each gender
sns.barplot(x='Gender', y='Amount', data=sales_gen)

# Adding plot labels and title
plt.xlabel('Gender')
plt.ylabel('Total Sales Amount')
plt.title('Total Sales Amount by Gender')

# Displaying the plot
plt.show()
```

From the bar plot that represents the total sales amount by gender, 
it appears that the majority of the buyers in the dataset are females, and 
the total purchasing power of females is higher compared to men. 
This observation is based on the total sales amount, and it suggests that females might be making more purchases or spending more on average than males.

## **Age_Buckets**

```python
# Creating a count plot to visualize the distribution of individuals across different age buckets with gender breakdown
ax = sns.countplot(data=df, x='Age_Buckets', hue='Gender')

# Adding labels to the bars with the exact count of individuals in each category
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='bottom')

# Adding plot labels and title
plt.xlabel('Age Buckets')
plt.ylabel('Count')
plt.title('Age Buckets Distribution with Gender Breakdown')

# Displaying the plot
plt.show()
```

```python
# Grouping the DataFrame by 'Age Group', calculating the sum of 'Amount', and sorting the results
sales_age = df.groupby(['Age_Buckets'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Creating a bar plot to visualize the total sales amount for each 'Age Group'
sns.barplot(x='Age_Buckets', y='Amount', data=sales_age)

# Adding plot labels and title
plt.xlabel('Age Buckets')
plt.ylabel('Total Sales Amount')
plt.title('Total Sales Amount by Age Buckets')

# Displaying the plot
plt.show()
```

From the bar plot representing the total sales amount by age group, it is evident that the majority of buyers fall within the age group of 26-35 years and are predominantly female. This insightful observation points to a significant market segment where females in their late twenties to mid-thirties contribute significantly to the overall sales. Businesses can leverage this information to tailor their marketing strategies and product offerings to cater specifically to this age group and gender, potentially boosting sales and customer engagement in this key demographic.

## **State**

```python
# Grouping the DataFrame by 'State', calculating the sum of 'Orders', and sorting the results
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Setting the figure size for the plot
sns.set(rc={'figure.figsize':(15, 5)})

# Creating a bar plot to visualize the total number of orders from the top 10 states
sns.barplot(data=sales_state, x='State', y='Orders')

# Adding plot labels and title
plt.xlabel('State')
plt.ylabel('Total Orders')
plt.title('Total Number of Orders from Top 10 States')

# Displaying the plot
plt.show()
```

```python
# Grouping the DataFrame by 'State', calculating the sum of 'Amount', and sorting the results
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Setting the figure size for the plot
sns.set(rc={'figure.figsize': (15, 5)})

# Creating a bar plot to visualize the total sales amount from the top 10 states
sns.barplot(data=sales_state, x='State', y='Amount')

# Adding plot labels and title
plt.xlabel('State')
plt.ylabel('Total Sales Amount')
plt.title('Total Sales Amount from Top 10 States')

# Displaying the plot
plt.show()
```


From the provided bar plots, we observe an interesting and unexpected trend in the distribution of orders and total sales amount across the top 10 states:

Order Distribution: Surprisingly, Uttar Pradesh (U.P.) takes the lead in terms of the number of orders, followed by Maharashtra and Karnataka, which rank second and third, respectively. This unexpected finding highlights the substantial volume of orders placed from these states.

Total Sales Amount: In contrast, when considering the total sales amount, U.P. takes the top position, with Karnataka following closely behind in second place. Maharashtra, which ranked second in order distribution, comes in third place in terms of the total sales amount.

This disparity between the order distribution and total sales amount showcases the significance of Uttar Pradesh and Karnataka in contributing both high order volume and substantial sales revenue. This intriguing insight can guide businesses to explore further into the market dynamics of these states, aiming to leverage their potential for increased revenue and growth. Moreover, it underlines the importance of considering both order count and total sales amount when formulating strategies to target specific regions.

## **Marital_Status**

```python
# Creating a count plot to visualize the distribution of individuals based on 'Marital_Status'
ax = sns.countplot(data=df, x='Marital_Status')

# Getting the total count of individuals for each category
total_counts = len(df['Marital_Status'])

# Adding labels to the bars with the exact count of individuals in each category
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='bottom')

# Adding plot labels and title
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.title(f'Distribution of Individuals by Marital Status (Total: {total_counts})')

# Displaying the plot
plt.show()
```

```python
# Grouping the DataFrame by 'Marital_Status' and 'Gender', calculating the sum of 'Amount', and sorting the results
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Setting the figure size for the plot
sns.set(rc={'figure.figsize': (6, 5)})

# Creating a bar plot to visualize the total sales amount for each 'Marital_Status' with gender breakdown
sns.barplot(data=sales_state, x='Marital_Status', y='Amount', hue='Gender')

# Adding plot labels and title
plt.xlabel('Marital Status')
plt.ylabel('Total Sales Amount')
plt.title('Total Sales Amount by Marital Status with Gender Breakdown')

# Displaying the plot
plt.show()
```

From the bar plot representing the total sales amount by 'Marital_Status' with a gender breakdown, it is evident that the majority of buyers are married women, and they exhibit a high purchasing power. This insightful observation points to a significant market segment where married women play a crucial role in contributing to the overall sales.

## **Occupation**

```python
# Setting the figure size for the plot
sns.set(rc={'figure.figsize': (20, 5)})

# Creating a count plot to visualize the distribution of individuals based on 'Occupation'
ax = sns.countplot(data=df, x='Occupation')

# Getting the total count of individuals for each category
total_counts = len(df['Occupation'])

# Adding labels to the bars with the exact count of individuals in each category
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='bottom')

# Adding plot labels and title
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.title(f'Distribution of Individuals by Occupation (Total: {total_counts})')

# Displaying the plot
plt.show()
```

```python

# Grouping the DataFrame by 'Occupation', calculating the sum of 'Amount', and sorting the results
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Setting the figure size for the plot
sns.set(rc={'figure.figsize': (20, 5)})

# Creating a bar plot to visualize the total sales amount for each 'Occupation'
sns.barplot(data=sales_state, x='Occupation', y='Amount')

# Adding plot labels and title
plt.xlabel('Occupation')
plt.ylabel('Total Sales Amount')
plt.title('Total Sales Amount by Occupation')

# Displaying the plot
plt.show()
```

This insightful observation points to a significant market segment where professionals from these occupational fields contribute prominently to the overall sales.

## **Product_Category**

```python
# Setting the figure size for the plot
sns.set(rc={'figure.figsize': (20, 5)})

# Creating a count plot to visualize the distribution of products based on 'Product_Category'
ax = sns.countplot(data=df, x='Product_Category')

# Getting the total count of products in each category
total_counts = len(df['Product_Category'])

# Adding labels to the bars with the exact count of products in each category
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='bottom')

# Adding plot labels and title
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.title(f'Distribution of Products by Product Category (Total: {total_counts})')

# Adjusting space between 'Product_Category' labels
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

# Displaying the plot
plt.show()
```

```python
# Grouping the DataFrame by 'Product_Category', calculating the sum of 'Amount', and sorting the results
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Setting the figure size for the plot
sns.set(rc={'figure.figsize': (20, 5)})

# Creating a bar plot to visualize the top 10 product categories by total sales amount
sns.barplot(data=sales_state, x='Product_Category', y='Amount')

# Adding plot labels and title
plt.xlabel('Product Category')
plt.ylabel('Total Sales Amount')
plt.title('Top 10 Product Categories by Total Sales Amount')

# Displaying the plot
plt.show()
```

Indeed, from the bar plot representing the top 10 product categories by total sales amount, it is evident that the majority of the sold products belong to the Food, Clothing, and Electronics categories. This insightful observation highlights the significance of these product categories in contributing substantially to the overall sales.

The visualization showcases the dominance of the Food, Clothing, and Electronics categories in terms of total sales amount, indicating their popularity and strong consumer demand. Businesses can leverage this information to focus on these high-performing product categories, optimize their inventory management, and tailor marketing strategies to further boost sales and customer engagement.

## **Product_ID**

```python
# Grouping the DataFrame by 'Product_ID', calculating the sum of 'Orders', and sorting the results
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Setting the figure size for the plot
sns.set(rc={'figure.figsize': (20, 5)})

# Creating a bar plot to visualize the top 10 products by the total number of orders
sns.barplot(data=sales_state, x='Product_ID', y='Orders')

# Adding plot labels and title
plt.xlabel('Product ID')
plt.ylabel('Total Orders')
plt.title('Top 10 Products by Total Number of Orders')

# Displaying the plot
plt.show()
```

```python
# Grouping the DataFrame by 'Product_ID', calculating the sum of 'Orders', and finding the top 10 products by order count
top_10_products = df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False)

# Creating a bar plot to visualize the top 10 most sold products
fig1, ax1 = plt.subplots(figsize=(12, 7))
top_10_products.plot(kind='bar', ax=ax1)

# Adding plot labels and title
plt.xlabel('Product ID')
plt.ylabel('Total Orders')
plt.title('Top 10 Most Sold Products by Total Number of Orders')

# Displaying the plot
plt.show()
```
